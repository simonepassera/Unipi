# Approfondimenti: Scipy, Bernoulli trials

* distribuzione di probabilità di una variabile casuale che vale $1$ con probabilità $p$  e $0$ con probabilità $q=1-p$
* corrisponde a quegli esperimenti, il cui output è una risposta sì/no
* adatta per rappresentare il lancio di una moneta in cui si ha successo (testa o croce) con probabilità $p$ e fallimento con probabilità $1-p$


* **fair coin** $p=0.5$, **unfair coin** $p \neq 0.5$

# Approfondimenti: Bernoulli trials

```python
from scipy.stats import bernoulli
# p=0.5 i.e. moneta "equa"
print(bernoulli.rvs(p=0.5,size=10))
```

```python
# p=0.2 i.e. più Croce (0) che Testa  (1)
print(bernoulli.rvs(p=0.2,size=20))
```

```python
# p=0.8 i.e. more Testa (1) che croce (0)
print(bernoulli.rvs(p=0.8,size=20))
```

# Ex 6: Lanci ripetuti di una moneta "truccata"

```python
import numpy as np
np.random.seed(0)
print("Simuliano un lancio della moneta truccata")
coin_flip = np.random.binomial(1, 0.7)
print(coin_flip)
print(f"E' uscita {'Testa' if coin_flip == 1 else 'Coda'}.")

print("\nSimuliamo 10 lanci della moneta")
number_coin_flips = 10
head_count = np.random.binomial(number_coin_flips, 0.7)
print(head_count)
print((f"{head_count} Testa su " 
       f"{number_coin_flips} lanci della moneta truccata"))
```

# Ex 6: Lanci ripetuti  di una moneta "truccata"

```python
np.random.seed(0)
head_count = np.random.binomial(1000, 0.7)
frequency = head_count / 1000
print(f"La percentuale di Testa è  {frequency}")
```

* La percentuale  è di .03 unità più piccola della reale probabilità di "Testa" che noi sappiamo essere 0.7

# Ex 6: Valutare la frequenza su più esperimenti


```python
np.random.seed(0)
assert np.random.binomial(1000, 0.7) / 1000 == 0.697
for i in range(1, 6):
    head_count = np.random.binomial(1000, 0.7)
    frequency = head_count / 1000
    print(f"La percentuale alla iterazione {i} è {frequency}")
    if frequency == 0.7:
        print("La percentuale è uguale alla probabilità!")
        
```

* la percentuale dell'evento "Testa" fluttua intorno al valore reale 0.7
* solo un campione di 1000 lanci ha prodotto un valore uguale alla probabilità corretta (0.7)

#  Ex 6: calcolare l'intervallo di confidenza

* vediamo adesso le cose da un altro punto di vista
* ci è stata consegnata una moneta che è "truccata", ma ma noi non lo sappiamo
* lanciamo 1000 volte la moneta e calcoliamo la probabilità di ottenere testa
* la probabilità fluttua intorno a 0.7, ma come possiamo concludere che la probabilità effettiva di testa (e quindi il bias) sia effettivamente di 0.7?
* necessario valutare un **intervallo di confidenza**
* proviamo ad aumentare  ulteriormente il numero dei lanci

#  Ex 6: Effettuare più esperimenti 

```python
np.random.seed(0)
head_count_list =[np.random.binomial(1000, 0.7) for _ in range(500)]
print (head_count_list[:20])
```

* in alternativa utilizzare direttamente il costrutto **array** di **numpy**

```python
np.random.seed(0)
head_count_array = np.random.binomial(1000, 0.7, 500)
print (head_count_array[:20])
assert head_count_array.tolist() == head_count_list
new_array = np.array(head_count_list)
assert np.array_equal(new_array, head_count_array) == True
```

# Ex 6: Calcolare la frequenza: lavorare con gli array

```python
frequency_array = head_count_array / 1000
assert frequency_array.tolist() == [head_count / 1000
                                    for head_count in head_count_list]

print(frequency_array[:20])

min_freq = frequency_array.min()
max_freq = frequency_array.max()

print(f"Minimum frequency observed: {min_freq}")
print(f"Maximum frequency observed: {max_freq}")
print(f"Difference across frequency range: {max_freq - min_freq}")
```

# Ex 6: Scatter plot dei risultati

```python
from collections import defaultdict
import matplotlib.pyplot as plt

frequency_counts = defaultdict(int)
for frequency in frequency_array:
    frequency_counts[frequency] += 1

frequencies = list(frequency_counts.keys())
counts = [frequency_counts[freq] for freq in frequencies]
plt.scatter(frequencies, counts)
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.show()
```

* il diagramma fornisce solo una idea della distribuzione, alcuni punti sono sovrapposti nella figura
* un istogramma può rappresentare meglio la distribuzione

#  Ex 6: Binning dei risultati

* dividere l'intervallo di valori delle frequenza in N sottointervalli (bins) della stessa dimensione:
    * opzione "auto": suddivisione automatica
* ogni frequenza cade in uno solo di questi bins
* per ogni bin viene visualizzato il numero di elementi del dataset che cade in quel bin


```python
counts, bin_edges, _ = plt.hist(frequency_array, bins='auto', edgecolor='black')
plt.xlabel('Binned Frequency')
plt.ylabel('Count')
plt.show()
print(f"Number of Bins: {counts.size}")
bin_width = bin_edges[1] - bin_edges[0]
print(f"Bin width: {bin_width}")
```

# Ex 6: Binning dei risultati

* **plt.hist()**
    * calcola e visualizza un istogramma
* restituisce
    * counts: array con i valori di ogni bin
    * bin_edges: estremi sinistri di ogni bin + estremo destro dell'ultimo bin, come ultimo elemento
    * un vettore patches che serve di riferimento per settare proprietà di quel bin

# Ex 6: Analizzare un particolare bin

```python
def output_bin_coverage(i):
    count = int(counts[i])
    range_start, range_end = bin_edges[i], bin_edges[i+1]
    range_string = f"{range_start} - {range_end}"
    print((f"The bin for frequency range {range_string} contains " 
           f"{count} element{'' if count == 1 else 's'}"))

output_bin_coverage(counts.argmax())
    
output_bin_coverage(0)
output_bin_coverage(5)
```

# Ex 6: Trasformare un istogramma in una distribuzione di probabilità

* il parametro densità normalizza le altezze dei bins in modo che la somma dei bins sia =1
* l'istogramma risultante  è una approssimazione della funzione di densità della probabilità

```python
likelihoods, bin_edges, _ = plt.hist(frequency_array, bins='auto', edgecolor='black', density=True)
plt.xlabel('Binned Frequency')
plt.ylabel('Relative Likelihood')
plt.show()
assert likelihoods.sum() * bin_width == 1.0
```

# Ex 6:  Calcolare l'intervallo di confidenza


```python
index = likelihoods.argmax()
area = likelihoods[index] * bin_width
range_start, range_end = bin_edges[index], bin_edges[index+1]
range_string = f"{range_start} - {range_end}"
print(f"il valore  cade nell'intervallo {range_string} con probabilità {area}")
```

* likelihoods: vettore che riporta i valori contenuti nei diversi istogrammi
* argmax. metodo di numpy che restituisce l'indice del bin che contiene il valore massimo, considero la probabilità di quel bin
* la probabilità relativa a quell'intervallo è molto bassa
    * confidenza bassa su quel valore
* cosa fare? Considerando un intervallo più ampio la probabilità aumenta

# Ex 6: Calcolare l'intervallo di confidenza: aumentare ampiezza intervallo

```python
peak_index = likelihoods.argmax()
start_index, end_index = (peak_index - 1, peak_index + 2)
area = likelihoods[start_index: end_index + 1].sum() * bin_width
range_start, range_end = bin_edges[start_index], bin_edges[end_index]
range_string = f"{range_start} - {range_end}"
print(f"Il valore cade nell'intervallo {range_string} con probabilità {area}")
```

* considero gli intervalli adiacenti a quello massimo
* il valore cade nell'intervallo intervallo  0.6896875 - 0.704125  con probabilità 0.464
* valore ancora troppo basso: in genere accettabile un intervallo di confidenza al 95%
* espandiamo gli intervalli finchè si raggiunge questo valore....

# Ex 6: Calcolare l'intervallo di confidenza: calcolare confidenza al 95%

```python
def compute_high_confidence_interval(likelihoods, bin_width):
    peak_index = likelihoods.argmax()
    area = likelihoods[peak_index] * bin_width
    start_index, end_index = peak_index, peak_index + 1
    while area < 0.95:
        if start_index > 0:
            start_index -= 1
        if end_index < likelihoods.size - 1:
            end_index += 1

        area = likelihoods[start_index: end_index + 1].sum() * bin_width
    
    range_start, range_end = bin_edges[start_index], bin_edges[end_index]
    range_string = f"{range_start:.6f} - {range_end:.6f}"
    print((f"L'intervallo {range_string} rappresenta un "
           f"{100 * area:.2f}% intervallo di confidenza"))
    return start_index, end_index

compute_high_confidence_interval(likelihoods, bin_width)
```

* La probabilità di successo cade tra  0.670 and 0.723, con un intervallo di confidenza del 95%
* obiettivo: diminuire l'intervallo mantenendo la stessa confidenza

# Ex 6: Calcolare l'intervallo di confidenza: aumentare numero esperimenti


```python
np.random.seed(0)
head_count_array = np.random.binomial(1000, 0.7, 100000)
frequency_array = head_count_array / 1000
assert frequency_array.size == 100000
likelihoods, bin_edges, patches = plt.hist(frequency_array, bins='auto', 
                                           edgecolor='black', density=True)
bin_width = bin_edges[1] - bin_edges[0]
start_index, end_index = compute_high_confidence_interval(likelihoods, 
                                                          bin_width)

for i in range(start_index, end_index):
      patches[i].set_facecolor('yellow')
plt.xlabel('Binned Frequency')
plt.ylabel('Relative Likelihood')

plt.show()
```

# Ex 6: Calcolare l'intervallo di confidenza: aumentare il numero di lanci

```python
np.random.seed(0)
head_count_array = np.random.binomial(50000, 0.7, 100000)
frequency_array = head_count_array / 50000

likelihoods, bin_edges, patches = plt.hist(frequency_array, bins='auto',
                                           edgecolor='black', density=True)
bin_width = bin_edges[1] - bin_edges[0]
start_index, end_index = compute_high_confidence_interval(likelihoods, bin_width)

for i in range(start_index, end_index):
     patches[i].set_facecolor('yellow')
plt.xlabel('Binned Frequency')
plt.ylabel('Relative Likelihood')

plt.show()
```
